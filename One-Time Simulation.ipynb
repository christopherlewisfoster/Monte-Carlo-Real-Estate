### Import Relevant Libraries

```python
import numpy as np
import pandas as pd
import random

from montecarlo import create_sp_return_generator, create_housing_return_generator, last_entry, REAsset, StockAsset, one_sim

%matplotlib inline
```

### Purchase condo valued at 480,000 dollars over 20-year mortgage

```python
results, ROI = one_sim(400_000, 30, 30)
results[['Stock Asset: Market Value', 'Stock Asset: Book Value', 'Stock Asset: Accrued Cost']].plot()
results[['RE Asset: Yearly Payment', 'RE Asset: Mortgage Payment', 'RE Asset: Rent', 'RE Asset: Interest Payment']].plot()
print("\nROI: {}".format(ROI))
```
